```python
data = pd.read_csv("business_data1.csv",index_col=0)
data
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# FACTOR VAR: Brand,Size, Grade,Pipe Type,Color
#CONTINUOUS VAR: Quantity,Value,Gross_total, Cartage, Sales
```

```python
data["CARTAGE"] = data.groupby(["Date", "Customer_Name"])["CARTAGE"].ffill()
```

```python
#cartage yes: 1 ; no: 0
```

```python
data["CARTAGE"] = np.where(data["CARTAGE"].isna(), 0, 1) 
```

```python
data = data[data["Gross_Total"] == 0]
```

```python
data = data.drop(columns="Gross_Total")
data = data.drop(columns="SALES")
```

```python
data["Color"] = data["Color"].replace({np.nan:"Default"})
```

```python
from scipy import stats
```

```python
dc=data1[data1['Brand']=='COMET']
dr=data1[data1['Brand']=='RSC']
```

```python
# h0: mu comet> mu RSC    mu comet-mu RSC>=0 
# h1: mu comet< mu RSC    mu comet-mu RSC<=0 
stats.ttest_ind(dc['Quantity'],dr['Quantity'],alternative='less')
## we reject the null hypo.  RSC has more impact on Quantity
```

```python
dt=data1[data1['Size']=='3/4']
do=data1[data1['Size']=='1']
```

```python
# Check for whitespace
data1['Size'] = data1['Size'].str.strip()

# Verify unique values
print(data1['Size'].unique())
```

```python
dt
```

```python
# h0: mu 3/4> mu 1    mu 3/4-mu 1>=0 
# h1: mu 3/4< mu 1    mu 3/4-mu 1<=0 
stats.ttest_ind(dt['Quantity'],do['Quantity'],alternative='less')
## we accept the null hypo. 3/4 inch has more impact on Quantity
```

```python

dh=data1[data1['Pipe_Type']=='HDPE']
ds=data1[data1['Pipe_Type']=='Steel']
```

```python
# h0:mu hpde-mu steel>=0 
# h1:mu hdpe-mu steel<=0 
stats.ttest_ind(dh['Quantity'],ds['Quantity'],alternative='less')
## we reject the null hypo. steel has more impact on Quantity
```

```python

doh=data1[(data1['Pipe_Type']=='HDPE')&(data1['Size']=='1*1/4')]
dth=data1[(data1['Pipe_Type']=='HDPE')&(data1['Size']=='2')]
```

```python
# h0:mu hpde 1*1/4-mu hdpe 2>=0 
# h1:mu hdpe 1*1/4-mu hdpe 2<=0 
stats.ttest_ind(doh['Quantity'],dth['Quantity'],alternative='less')
## we reject the null hypo. dhpe 2 has more impact on Quantity
```

```python
dc=data1[data1['Brand']=='COMET']
db=data1[data1['Brand']=='BLACK']
```

```python
# h0: mu comet> mu black    mu comet-mu black>=0 
# h1: mu comet< mu black    mu comet-mu black<=0 
stats.ttest_ind(dc['Value'],db['Value'],alternative='less')
## we reject the null hypo.  Black has more impact on Value
```

```python
done=data1[data1['Size']=='1*1/4']
df=data1[data1['Size']=='5']
```

```python
# h0: mu 1*1/4> mu 5    mu 1*1/4-mu 5>=0 
# h1: mu 1*1/4< mu 5   mu 1*1/4-mu 5<=0 
stats.ttest_ind(done['Value'],df['Value'],alternative='less')
## we rejct the null hypo. 5 inch has more impact on value
```

```python
#pipe 
# h0:mu hpde-mu steel>=0 
# h1:mu hdpe-mu steel<=0 
stats.ttest_ind(dh['Value'],ds['Value'],alternative='less')
## we reject the null hypo. steel has more impact on value
```

```python
dho=data1[(data1['Pipe_Type']=='HDPE')&(data1['Size']=='1/2')]
ds5=data1[(data1['Pipe_Type']=='Steel')&(data1['Size']=='5')]
```

```python
# h0:mu hpde 1/2-mu steel 5>=0 
# h1:mu hdpe 1/2-mu steel 5<=0 
stats.ttest_ind(dho['Value'],ds5['Value'],alternative='less')
## we reject the null hypo. steel 5 inch has more impact on value
```

```python
dbf=data1[(data1['Brand']=='BLACK')&(data1['Size']=='5')]
dro=data1[(data1['Brand']=='RIDDHI')&(data1['Size']=='1/2')]
```

```python
# h0:mu black 5-mu riddhi 1/2>=0 
# h1:mu black 5-mu riddhi 1/2<=0 
stats.ttest_ind(dbf['Value'],dro['Value'],alternative='less')
## we accept the null hypo. black 5 inch has more impact on value than riddhi 1/2
```
